```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Leemos el archivo 
df = pd.read_csv('C:/Users/PC/Desktop/proyecto/Data/movies.csv')
```

```python
#Vemos que se haya cargado 
df.head(1)
```

```python
#Revisamos todos los valores nulos en las columnas 
df.isnull().sum()
```

```python
#Hacemos una descripcion estadistica de las siguientes columnas 
df[['revenue', 'runtime', 'popularity', 'vote_average', 'release_year']].describe()
```

```python
#Revisamos los valores unicos en las siguientes columnas
for col in ['belongs_to_collection', 'genres', 'production_companies', 'production_countries']:
    print(f"{col} tiene {df[col].nunique()} valores únicos")
    print(df[col].value_counts().head())
    print("\n")
```

```python
#Creamos histomagras para las 2 siguientes columnas
sns.histplot(df["vote_average"], bins=10)
plt.show()

sns.histplot(df["release_year"], bins=10)
plt.show()
```

```python
# Utiliza la función pairplot de Seaborn para generar un gráfico de pares
sns.pairplot(df[["popularity","vote_average"]])
plt.show()
```

```python
#Creamos una matris de correlacion 
corr_matrix = df[["popularity", "revenue", "runtime", "vote_average", "release_year"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```
